# Sample Calculations
## Sulfuric Acid Molarity Calculations
Sulfuric Acid Concentration = $96.7\% w/w$  
Molar Mass = $98.08 \frac{g}{mol}$  
$ρ_{H_2SO_4} = 1.84 \frac{g}{mL}$   $ρ_{H_2O} = 1.0\frac{g}{mL}$ 

$\left(96.7\frac{\text{g} H_2SO_4}{100g} \text{solution}\right) \times \left(1.84\frac{\text{g} H_2SO_4}{mL}\right) = 1.78\frac{\text{g}}{\text{mL}} \text{solution}\times \left(1000\frac{\text{mL}}{\text{L}}\right) \times \left(\frac{1\text{mol} H_2SO+4}{98.08\text{g}}\right) = 18.1 \frac{\text{mol}}{\text{L}}  = 18.1 \text{M}$ 

### Diluting to Desired Molarity
$M_1V_1 = M_2V_2$  
Desired Molarity = $0.4 M$  
Desired Volume = $0.25L$  
$H_2SO_4$ to add = $\dfrac{0.4M * 0.25 L}{18.1 M} \times \dfrac{1000 mL}{1 L}$ = 5.52 mL $H_2SO_4$  
Remaining water = 250mL - 5.52 mL = 244.48 mL

## Calculation of Concentration from Absorbance
Beer Lambert’s Law  
$A = ε c l$  
Here $ε = 3.34 \frac{\text{mol}}{\text{L}}\times \text{cm}$ (average taken from previous groups),  $l = 1\text{cm}$, 
 c = $A\frac{\left(xxxx\right)}{3.34 \times 1}$

## Thomas Model
$ln\left(\dfrac{C_F}{C}-1\right) = \dfrac{k_{Th}\;q_F\;m}{Q} - k_{Th} C_F t$  

To determine $k_{Th}$ in these calcultaions, plot $ln\left(\dfrac{C_F}{C}-1\right)$ vs time.  The slope is equal to $-k_{Th}\cdot C_F$  
Therefore, it follows that $k_{Th} = \dfrac{-\text{slope}}{C_F}$

$q_F$ is determined from the y-intercept of the same plot. The y intercept is equal to $\dfrac{k_{Th}\cdot q_F \cdot m}{Q}$

Therefore, $q_F = \dfrac{\text{y-intercept}\cdot Q}{m \cdot k_{Th}}$

## Plotting the thomas model plot before and after regeneration

```python
ideal3['Thomas'] = np.log((CF/ideal3['Concentration'])-1)
regen1_3['Thomas'] = np.log((CF/regen1_3['Concentration'])-1)
ax = ideal3.plot.scatter(x='time_s',y='Thomas',label='Breakthrough 1',color='C0')
regen1_3.plot.scatter(x='time_s',y='Thomas',label='After Regeneration 1',ax=ax,color='C1')
ax.set_xlabel('time(min)')
ax.set_ylabel('$ln(\dfrac{C_F}{C} -1)$')

title3 = "Thomas Model Curve Comparing before and After \n regeneration at Flowrate = {} mL/min".format(round(Q3,3))
ax.set_title(title3)

import scipy as sc
from scipy import stats

slopeI3, interceptI3, r_valueI3, p_valueI3, std_errI3 = stats.linregress(ideal3['time_s'],ideal3['Thomas'])
slopeR3, interceptR3, r_valueR3, p_valueR3, std_errR3 = stats.linregress(regen1_3['time_s'],regen1_3['Thomas'])
ideal3['ThomasPredict'] = ideal3['time_s']*slopeI3 + interceptI3
regen1_3['ThomasPredict'] = regen1_3['time_s']*slopeR3 + interceptR3
ideal3.plot(x='time_s',y='ThomasPredict', ax=ax,label='Ideal',color='C0')
regen1_3.plot(x='time_s',y='ThomasPredict', ax=ax,label='Regenerated',color='C1')

print([[slopeI3, interceptI3, (r_valueI3**2),p_valueI3,std_errI3],[slopeR3, interceptR3, (r_valueR3**2),p_valueR3,std_errR3]])
```

### Reporting slope and intercept for Thomas Model at FR3

```python
print("For Ideal Case: \nSlope = {}\n y- intercept = {} \n R^2 value = {}".format(slopeI3, interceptI3,r_valueI3**2))
print("After 1 regeneration: \nSlope = {}\n y- intercept = {} \n R^2 value = {}".format(slopeR3, interceptR3,r_valueR3**2))
```

### Calculating $q_F$ and $k_{Th}$ from Thomas model
Below calculates the Thomas Model parameters, $q_F$ and $k_{Th}$, from the calculated slope and intercept, and measured values, such as the flowrate, and mass of resin.

```python
kthI3 = -slopeI3/CF
qFI3 = (interceptI3 * Q3)/(m3 * kthI3)

print(m3)

kthR3 = -slopeR3/CF
qFR3 = (interceptR3 * Q3)/(m3 * kthR3)

print("Flowrate = {} mL/min \n Ideal case \n \tkth: {} \n\tqF: {}".format(round(Q3,3),round(kthI3,3), round(qFI3,3)))
print("After 1 regeneration\n \tkth: {} \n\tqF: {}".format(round(kthR3,3), round(qFR3,3)))
print("Efficiency = {}".format(round((qFR3*100/qFI3),3)))
```

### Calculating $q_F$ and $k_{Th}$ from Thomas model
